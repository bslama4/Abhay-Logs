```python
import pandas as pd
from btalib import RSI, BBANDS, EMA, SMA, LINEARREG_SLOPE
from matplotlib import pyplot as plt

# Five_minute = pd.read_csv("df_hdfc_5minute.csv")  # read the csv file of five minute chart
# Five_minute = pd.read_csv("df_5_minute Small.csv")  # read the csv file of five minute chart
Five_minute = pd.read_csv("E:\\Kite_INFY_1D (1).csv")  # read the csv file of five minute chart

''' Defining the RSI, EMA, SMA, BB for 5 minute chart  '''

up, mid, low = BBANDS(Five_minute['close'], timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)
rsi = RSI(Five_minute['close'], timeperiod=14)
ema = EMA(Five_minute['close'], 5)
sma = SMA(Five_minute['close'], 5)
slope = LINEARREG_SLOPE(mid, timeperiod=20)
up = up.dropna()
mid = mid.dropna()
low = low.dropna()

Five_minute['RSI'] = rsi.values
Five_minute['SMA'] = sma.values
Five_minute['EMA'] = ema.values
Five_minute['SLOPE'] = slope.values

Five_minute = Five_minute.dropna()
Five_minute = Five_minute.reset_index().drop(['index'], axis=1)

# One_minute = pd.read_csv("df_hdfc_1minute.csv")  # read the csv file of one minute chart
# One_minute = pd.read_csv("df_1_minute Small.csv")  # read the csv file of five minute chart
One_minute = pd.read_csv("E:\\Kite_INFY_1D (1).csv")  # read the csv file of five minute chart

''' Defining the RSI, EMA, SMA, BB for 1 minute chart  '''

up, mid, low = BBANDS(One_minute['close'], timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)
rsi = RSI(One_minute['close'], timeperiod=14)
ema = EMA(One_minute['close'], 5)
sma = SMA(One_minute['close'], 5)
slope = LINEARREG_SLOPE(mid, timeperiod=20)
up = up.dropna()
mid = mid.dropna()
low = low.dropna()

One_minute['RSI'] = rsi.values
One_minute['SMA'] = sma.values
One_minute['EMA'] = ema.values
One_minute['SLOPE'] = slope.values

One_minute = One_minute.dropna()
One_minute = One_minute.reset_index().drop(['index'], axis=1)

''' If we want to plot the data we can use these lines of code '''

#fig, ax0 = plt.subplots()
#ax0.plot(One_minute['date'], One_minute['close'], label='Close')
#ax0.set_xlabel('Date')
#ax0.set_ylabel('Close')

buy_threshold = 0.2
sell_threshold = -buy_threshold  # threshold for slope of bollinger bands


# ---------------------------------------------------------
# If slope > 0.2 -> We are only buying 
# If slope < -0.2 -> We are only selling

# If slope BETWEEN 0.2 and -0.2 AND certain conditions are met, we are in the 'sideways trading' logic AND we are BUYING
# If slope BETWEEN 0.2 and -0.2 AND certain different conditions are met, we are in the 'sideways trading' logic AND we are SELLING
# ---------------------------------------------------------



printEntranceInfo = True
printTradeInfo = True

count_upward = 0  # this variable is used in uptrending chart for increasing the stop_loss
count_downward = 0  # this variable is used in downtrending chart for increasing the stop_loss
limit = ''
previous_high = Five_minute.iloc[0]['high']
previous_low = Five_minute.iloc[0]['low']
prev_row_five = Five_minute.iloc[0]
Five_minute = Five_minute.iloc[1:]

winning_trade = 0
loosing_trade = 0
win_count = 0
lose_count = 0

entranceDF = pd.DataFrame(columns = ['Entrance_time', 'Side'])
exitDF = pd.DataFrame(columns = ['Exit_time', 'Side', 'P/L'])


for i, j in Five_minute.iterrows():  # we start iterating over the 5 minute chart data
    if j['date'] <= limit:
        continue
    # ************************UPWARD MOVING******************************
    if (j['SLOPE'] > buy_threshold):

        if (     (prev_row_five['EMA'] < prev_row_five["SMA"]) and (j['EMA'] > j["SMA"])) and j['RSI'] > 50:

            One_minute = One_minute[(One_minute.date >= j['date'])]  # getting the 1 minute data from when the condition satisfy(RSI, SMA/EMA, BB) for 5 minute chart
            prev_row_one = One_minute.iloc[0]
            One_minute = One_minute.iloc[1:]

            for l, k in One_minute.iterrows():

                if ((prev_row_one['EMA'] < prev_row_one["SMA"]) and (k['EMA'] > k["SMA"])) and k['RSI'] > 50 and k[
                    'SLOPE'] > buy_threshold:

                    if (printEntranceInfo):
                        print('SIGNAL INFO: prev_row[EMA]: ', prev_row['EMA'], ' prev_row[SMA]: ', prev_row['SMA'], ' k[EMA]: ', k['EMA'], ' k[SMA]: ', k['SMA'])
                        print('SIGNAL INFO: k[RSI]: ', k['RSI'], ' k[SLOPE]: ', k['SLOPE'], ' sell_threshold: ', sell_threshold, ' buy_threshold: ', buy_threshold)

                    stop_loss = previous_low

                    size_of_stop_loss = abs(stop_loss - k['close'])
                    target_price_1 = k['close'] + size_of_stop_loss

                    if (printTradeInfo):
                        print('TRADE INFO: k[close]: ', k['close'], ' stop_loss: ', stop_loss, ' size_of_stop_loss: ', size_of_stop_loss, ' target_price_1: ', target_price_1)

                    print("Enter, BUY NOW : ", k['date'], k['close'])

                    entranceDF.loc[len(entranceDF.index)] = [k['date'], 'BUY']
                    # Use this in the reset of the stop loss
                    entrancePriceLong = k['close']

                    temp_df = One_minute[One_minute.date > k['date']]
                    temp_prev = temp_df.iloc[0]
                    temp_df = temp_df.iloc[1:]

                    for a, z in temp_df.iterrows():  # this loop is for checking if market moves farther in our favor.

                        # Exiting condition
                        if z['low'] <= stop_loss:
                            print("EXIT, SELL NOW : ", z['date'], stop_loss)
                            limit = z['date']  # this is for prevention of iterating over same data again, in next iteration of 5 minute data we will start from this time.
                            print("UPWARD STOP LONG\n")
                            count_upward = 0
                            print("Looser rate : ", abs(entrancePriceLong-stop_loss)*100/entrancePriceLong)
                            print("Looser factor : ", stop_loss/entrancePriceLong)
                            loosing_trade += (stop_loss-entrancePriceLong)
                            lose_count += 1
                            exitDF.loc[len(exitDF.index)] = [z['date'], 'SELL',stop_loss-entrancePriceLong]
                            break

                        if z['high'] > target_price_1:  # if high value hits our target price 1 then change stop_loss value
                            count_upward += 1
                            if count_upward == 1:  # if first time crossing the target price...
                                stop_loss = entrancePriceLong

                            if z['close'] > (target_price_1 + size_of_stop_loss):
                                stop_loss = target_price_1  # setting the stop loss to target price 1

                        if (temp_prev['EMA'] > temp_prev["SMA"]) and (z['EMA'] < z["SMA"]):
                            # ax0.scatter(x=a, y=One_minute.loc[a, 'close'], color='green')
                            print("EXIT, SELL NOW : ", z['date'], z['close'])
                            limit = z['date']
                            print("UPWARD ALT EXIT LONG\n")
                            count_upward = 0
                            print("Winner rate: ", abs(entrancePriceLong-z['close'])*100/entrancePriceLong)
                            print("Winner factor : ", z['close']/entrancePriceLong)
                            winning_trade += (z['close']-entrancePriceLong)
                            win_count += 1
                            exitDF.loc[len(exitDF.index)] = [z['date'], 'SELL', entrancePriceLong-z['close']]
                            break

                        temp_prev = k

                    else:
                        continue
                    break  # break from both loops

                previous_high = k['high']
                prev_row_one = k
                
            prev_row_five = j

    # ************************DOWNWARD MOVING******************************
    elif (j['SLOPE'] < sell_threshold):

        if ((prev_row_five['EMA'] > prev_row_five["SMA"]) and (j['EMA'] < j["SMA"])) and j['RSI'] < 50:

            ''' Comments for this sections are same as UPWARD MOVING '''
            One_minute = One_minute[(One_minute.date >= j['date'])]

            for l, k in One_minute.iterrows():
                if ((prev_row['EMA'] > prev_row["SMA"]) and (k['EMA'] < k["SMA"])) and k['RSI'] < 50 and k['SLOPE'] < sell_threshold:

                    if (printEntranceInfo):
                        print('SIGNAL INFO: prev_row[EMA]: ', prev_row['EMA'], ' prev_row[SMA]: ', prev_row['SMA'], ' k[EMA]: ', k['EMA'], ' k[SMA]: ', k['SMA'])
                        print('SIGNAL INFO: k[RSI]: ', k['RSI'], ' k[SLOPE]: ', k['SLOPE'], ' sell_threshold: ', sell_threshold, ' buy_threshold: ', buy_threshold)

                    stop_loss = previous_high

                    size_of_stop_loss = abs(stop_loss - k['close'])
                    target_price_1 = k['close'] - size_of_stop_loss
                    

                    if (printTradeInfo):
                        print('TRADE INFO: k[close]: ', k['close'], ' stop_loss: ', stop_loss, ' size_of_stop_loss: ', size_of_stop_loss, ' target_price_1: ', target_price_1)

                    print("Enter, SELL NOW : ", k['date'], k['close'])

                    entranceDF.loc[len(entranceDF.index)] = [k['date'], 'SELL']
                    
                    # Use this in the reset of the stop loss
                    entrancePriceShort = k['close']

                    temp_df = One_minute[One_minute.date > k['date']]
                    temp_prev = temp_df.iloc[0]
                    temp_df = temp_df.iloc[1:]

                    for a, z in temp_df.iterrows():

                        if z['high'] >= stop_loss:
                            # ax0.scatter(x=a, y=One_minute.loc[a, 'close'], color='red')
                            print("EXIT, BUY NOW : ", z['date'], stop_loss)
                            limit = z['date']
                            print("DOWNWARD STOP SHORT\n")
                            count_downward = 0
                            print("Looser rate : ", abs(entrancePriceShort-stop_loss)*100/entrancePriceShort)
                            print("Looser factor : ", stop_loss/entrancePriceShort)
                            loosing_trade += (stop_loss-entrancePriceShort)
                            lose_count += 1
                            exitDF.loc[len(exitDF.index)] = [z['date'], 'BUY',stop_loss-entrancePriceShort]
                            break

                        if z['low'] < target_price_1:
                            count_downward += 1
                            if count_downward == 1:
                                stop_loss = entrancePriceShort
                            if z['close'] >= target_price_1 + size_of_stop_loss:
                                stop_loss = target_price_1

                        if (temp_prev['EMA'] > temp_prev["SMA"]) and (z['EMA'] < z["SMA"]):
                            # ax0.scatter(x=a, y=One_minute.loc[a, 'close'], color='red')
                            print("EXIT, BUY NOW : ", z['date'], z['close'])
                            limit = z['date']
                            print("DOWNWARD ALT EXIT SHORT\n")
                            count_downward = 0
                            print("Winner rate: ", abs(entrancePriceShort-z['close'])*100/entrancePriceShort)
                            print("Winner factor : ", z['close']/entrancePriceShort)
                            winning_trade += (z['close']-entrancePriceShort)
                            win_count += 1
                            exitDF.loc[len(exitDF.index)] = [z['date'], 'BUY',entrancePriceShort-z['close']]
                            break

                        temp_prev = z

                    else:
                        continue
                    break  # break from both loops
                previous_high = k['high']
                prev_row_one = k
                
            prev_row_five = j
            # ************************SIDEWAYS MOVING*****************************
    else:
        One_minute = One_minute[(One_minute.date >= j['date'])]
        prev_row = One_minute.iloc[0]
        One_minute = One_minute.iloc[1:]
        for l, k in One_minute.iterrows():

            if ((prev_row['EMA'] > prev_row["SMA"]) and (k['EMA'] < k["SMA"]) and k['RSI'] < 50 and (k['SLOPE'] > sell_threshold) and (k['SLOPE'] < buy_threshold)):

                if (printEntranceInfo):
                    print('SIGNAL INFO: prev_row[EMA]: ', prev_row['EMA'], ' prev_row[SMA]: ', prev_row['SMA'], ' k[EMA]: ', k['EMA'], ' k[SMA]: ', k['SMA'])
                    print('SIGNAL INFO: k[RSI]: ', k['RSI'], ' k[SLOPE]: ', k['SLOPE'], ' sell_threshold: ', sell_threshold, ' buy_threshold: ', buy_threshold)

                stop_loss = previous_high

                size_of_stop_loss = abs(k['close'] - stop_loss)

                target_price = k['close'] - size_of_stop_loss

                if (printTradeInfo):
                    print('TRADE INFO: k[close]: ', k['close'], ' stop_loss: ', stop_loss, ' size_of_stop_loss: ', size_of_stop_loss, ' target_price: ', target_price)

                print("Enter, SELL NOW: ", k['date'], k['close'])

                entranceDF.loc[len(entranceDF.index)] = [k['date'], 'SELL']

                temp_df = One_minute[One_minute.date > k['date']]

                for a, z in temp_df.iterrows():

                    if z['high'] >= stop_loss:
                        # ax0.scatter(x=a, y=One_minute.loc[a, 'close'], color='red')
                        print("EXIT, BUY NOW : ", z['date'], stop_loss)
                        limit = z['date']
                        print("SIDEWAYS STOP SHORT\n")
                        print("Looser rate : ", abs(k['close']-stop_loss)*100/k['close'])
                        print("Looser factor : ", stop_loss/k['close'])
                        loosing_trade += abs(stop_loss - k['close'])
                        lose_count += 1
                        exitDF.loc[len(exitDF.index)] = [z['date'], 'BUY',stop_loss-k['close']]
                        break

                    if z['low'] < target_price:
                        # ax0.scatter(x=a, y=One_minute.loc[a, 'close'], color='red')
                        print("EXIT, BUY NOW : ", z['date'], target_price)
                        limit = z['date']
                        print("SIDEWAYS TARGET SHORT\n")
                        print("Winner rate: ", abs(k['close']-target_price)*100/k['close'])
                        print("Winner factor : ", target_price/k['close'])
                        winning_trade += abs(target_price - k['close'])
                        win_count += 1
                        exitDF.loc[len(exitDF.index)] = [z['date'], 'BUY', k['close']-target_price]
                        break

                else:
                    continue
                break  # break from both loops

            elif ((prev_row['EMA'] < prev_row["SMA"]) and (k['EMA'] > k["SMA"]) and k['RSI'] > 50 and (k['SLOPE'] > sell_threshold) and (k['SLOPE'] < buy_threshold)):

                if (printEntranceInfo):
                    print('SIGNAL INFO: prev_row[EMA]: ', prev_row['EMA'], ' prev_row[SMA]: ', prev_row['SMA'], ' k[EMA]: ', k['EMA'], ' k[SMA]: ', k['SMA'])
                    print('SIGNAL INFO: k[RSI]: ', k['RSI'], ' k[SLOPE]: ', k['SLOPE'], ' sell_threshold: ', sell_threshold, ' buy_threshold: ', buy_threshold)

                stop_loss = previous_low

                size_of_stop_loss = abs(stop_loss - k['close'])
                target_price = k['open'] + size_of_stop_loss

                if (printTradeInfo):
                    print('TRADE INFO: k[close]: ', k['close'], ' stop_loss: ', stop_loss, ' size_of_stop_loss: ', size_of_stop_loss, ' target_price: ', target_price)

                print("Enter, BUY NOW: ", k['date'], k['close'])

                entranceDF.loc[len(entranceDF.index)] = [k['date'], 'BUY']

                temp_df = One_minute[One_minute.date > k['date']]

                for a, z in temp_df.iterrows():

                    if z['low'] <= stop_loss:
                        print("EXIT, SELL NOW : ", z['date'], stop_loss)
                        limit = z['date']
                        print("SIDEWAYS STOP LONG\n")

                        print("Looser rate : ", abs(k['open']-stop_loss)*100/k['open'])
                        print("Looser factor : ", stop_loss/k['open'])
                        loosing_trade += (stop_loss-k['open'])
                        lose_count += 1
                        exitDF.loc[len(exitDF.index)] = [z['date'], 'SELL',stop_loss-k['open']]
                        break

                    if z['high'] > target_price:
                        print("EXIT, SELL NOW : ", z['date'], target_price)
                        limit = z['date']
                        print("SIDEWAYS TARGET LONG\n")

                        print("Winner rate: ", abs(k['open']-target_price)*100/k['open'])
                        print("Winner factor : ", target_price/k['open'])
                        winning_trade += (target_price-k['open'])
                        win_count += 1
                        exitDF.loc[len(exitDF.index)] = [z['date'], 'SELL', target_price-k['open']]
                        break
                else:
                    continue
                break  # break from both loops

            previous_high = k['high']
            previous_low = k['low']

print("\n")
print("\n")
#fig.tight_layout()
#plt.show()
```
